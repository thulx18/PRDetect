```python
import json
import matplotlib
import matplotlib.pyplot as plt
```

```python
hc3human_len = []
hc3machine_len = []
gpt35human_len = []
gpt35machine_len = []
for file in ["hc3_train", "hc3_val", "hc3_test"]:
    with open(f"original_text/{file}.json", 'r', encoding='utf-8') as f:
        for line in f.readlines():
            line = json.loads(line)
            if line['label'] == "hc3human":
                hc3human_len.append(len(line['text'].strip().split()))
            else:
                hc3machine_len.append(len(line['text'].strip().split()))

for file in ["gpt3.5_mixed_train", "gpt3.5_mixed_val", "gpt3.5_mixed_test"]:
    with open(f"original_text/{file}.jsonl", 'r', encoding='utf-8') as f:
        for line in f.readlines():
            line = json.loads(line)
            if line['label'] == "human":
                gpt35human_len.append(len(line['article'].strip().split()))
            else:
                gpt35machine_len.append(len(line['article'].strip().split()))
```

```python
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.serif'] = ['Times New Roman']
fig, axs = plt.subplots(1, 2, figsize=(7, 3))
print(sum(hc3human_len) / len(hc3human_len))
print(sum(hc3machine_len) / len(hc3machine_len))
print((sum(hc3human_len)+sum(hc3machine_len)) / (len(hc3human_len)+len(hc3machine_len)))
max_len = min(max(hc3human_len), max(hc3machine_len)) * 1.5
L1 = []
L2 = []
for i in hc3human_len:
    if i <= max_len:
        L1.append(i)
for i in hc3machine_len:
    if i <= max_len:
        L2.append(i)
axs[0].hist(L1, bins=40, color="#1f77b4", alpha=0.6, label="human")
axs[0].hist(L2, bins=40, color="#ff7f0e", alpha=0.6, label="machine")
axs[0].set_title('HC3')

# axs[0].xlabel('Text Length')
# axs[0].ylabel('Frequency')
# axs[0].savefig("hc3.jpg")
# plt.show()

print(sum(gpt35human_len) / len(gpt35human_len))
print(sum(gpt35machine_len) / len(gpt35machine_len))
print((sum(gpt35human_len)+sum(gpt35machine_len)) / (len(gpt35human_len)+len(gpt35machine_len)))
max_len = min(max(gpt35human_len), max(gpt35machine_len)) * 1.5
L1 = []
L2 = []
for i in gpt35human_len:
    if i <= max_len:
        L1.append(i)
for i in gpt35machine_len:
    if i <= max_len:
        L2.append(i)
axs[1].hist(L1, bins=40, color="#1f77b4", alpha=0.6, label="human")
axs[1].hist(L2, bins=40, color="#ff7f0e", alpha=0.6, label="machine")
axs[1].set_title('GPT3.5-Mixed')
# plt.xlabel('Text Length')
# plt.ylabel('Frequency')
plt.legend()
fig.text(0.5, 0.01, 'Text Length', ha='center', fontsize=12)
fig.text(0.01, 0.5, 'Frequency', va='center', rotation='vertical', fontsize=12)
plt.savefig("len.pdf")
plt.show()
```
